# WEB-SCRAPPING(ACTIVE_STOCKS)

#### Importing necessary libraries

```python
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
url="https://finance.yahoo.com/markets/stocks/most-active/"
```

#### The purpose of this code to ask for permission for websites data,response other than[200] mean permission denied

```python
page=requests.get(url)
page
```

#### BeautifulSoup library is to parse HTML content and format it for easier readability,Prettify HTML content to make it more readable and visually appealing.

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

#### The code is used to locate and extract a table (specifically its body) from an HTML document using BeautifulSoup. Here’s a detailed explanation of each part of the process:

```python
table=soup.find("table")
table=soup.find("tbody")
table
```

#### The code is used to find all the rows in a table by searching for <tr> (table row) elements within a previously identified table or table body. Here’s a detailed explanation of the purpose of this line:

```python
rows=table.find_all("tr")
rows
```

#### The code is used to extract stock information from the rows of a table in a web page and store that information in a structured list. Here’s a detailed breakdown of each component and the overall purpose of the code:

```python
stock_info=[]
for i in rows:
    cols=i.find_all("td")
    T10=[P.text for P in cols]
    stock_info.append(T10)
print(stock_info)
```

```python
print(stock_info)
    
```

#### The code is used to extract and organize specific pieces of stock information from a previously created list, stock_info, and store each piece of information in separate lists for further analysis or processing. Here’s a detailed breakdown of each part of the process:

```python
Stock_Name=[]
Price=[]

Change=[]
Change_percentage=[]

Volume=[]
Avg_Volume=[]

Market_Cap=[] 
PE_Ratio=[]    

_52WeekChange_Percentage=[] 
for i in stock_info:

    Stock_Name.append(i[0])
    Price.append(i[1])

    Change.append(i[2])
    Change_percentage.append(i[3])

    Volume.append(i[4])
    Avg_Volume.append(i[5])

    Market_Cap.append(i[6])
    PE_Ratio.append(i[7])

    _52WeekChange_Percentage.append(i[8])
```

#### The code is used to create a structured dictionary to organize the stock information extracted from the previous steps. This dictionary will make it easier to manipulate, analyze, and possibly convert the data into a DataFrame for further processing or visualization. Here’s a detailed breakdown of the process:

```python
Data={"Stock_Name":Stock_Name,"Price":Price,"Change":Change,"Change_percentage":Change_percentage,"Volume":Volume,"Avg_Volume":Avg_Volume,"Market_Cap":Market_Cap,"PE_Ratio":PE_Ratio,"_52WeekChange_Percentage":_52WeekChange_Percentage}
Data
```

```python
stock_info=pd.DataFrame(Data)
```
